```python
matplotlib.rcParams['text.usetex'] = False
data = pd.read_csv('./data_gdp.csv', index_col='Country Name')
data=data.astype(float)
data.T.plot()
plt.legend(loc='upper left')
plt.xlabel('Year')
plt.ylabel('GDP per capita (current US$)')
```

```python
data = data.T
```

```python
growth_rate_brazil = np.exp(np.diff(np.log(data["Brazil"]))) - 1
growth_rate_USA = np.exp(np.diff(np.log(data["United States"]))) - 1
growth_rate_Canada = np.exp(np.diff(np.log(data["Canada"]))) - 1
```

```python
plt.bar(np.arange(len(growth_rate_USA)),growth_rate_USA)

plt.title("Brazil")

plt.xlabel('Years from first measurement')
plt.ylabel('Growth of GDP per capita ($)')

plt.show()
```

```python
plt.bar(np.arange(len(growth_rate_USA)),growth_rate_USA)
plt.title("United states")

plt.xlabel('Years from first measurement')
plt.ylabel('Growth of Growth of GDP per capita ($)')

plt.show()
```

```python
plt.bar(np.arange(len(growth_rate_Canada)),growth_rate_Canada)

plt.title("Canada")

plt.xlabel('Years from first measurement')
plt.ylabel('Growth of GDP per capita ($)')

plt.show()
```
